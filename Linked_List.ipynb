## 1. What is a Linked List?

A **Linked List** is a linear data structure where:
- Elements are stored in **nodes**
- Each node contains:
  - **Data**
  - **Reference (pointer)** to the next node

### Why Linked List?
In arrays:
- Size is fixed
- Insertion/deletion is costly

Linked Lists solve these problems.

### Representation
```
[data | next] ‚Üí [data | next] ‚Üí NULL
```



## 2. Node Structure (Basic Building Block)

Each node has:
- `data`
- `next` (address of next node)


````python

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
````

````python
class NODE:
    def __init__(self, data):
        self.data = data
        self.next = None


class LINKED_LIST:
    def __init__(self):
        self.head = None

    # -------- Traversal --------
    def traverse(self):
        temp = self.head
        size = 0
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
            size += 1
        print("NULL")
        return size

    # -------- Append (Insert at end) --------
    def append(self, data):
        new_node = NODE(data)

        # Empty list
        if self.head is None:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # -------- Prepend (Insert at beginning) --------
    def prepend(self, data):
        new_node = NODE(data)
        new_node.next = self.head
        self.head = new_node

    # -------- Insert at index (0-based) --------
    def insert(self, data, index):
        if index < 0:
            print("Invalid index")
            return

        # Insert at beginning
        if index == 0:
            self.prepend(data)
            return

        temp = self.head
        for _ in range(index - 1):
            if temp is None:
                print("Index out of range")
                return
            temp = temp.next

        if temp is None:
            print("Index out of range")
            return

        new_node = NODE(data)
        new_node.next = temp.next
        temp.next = new_node

    # -------- Delete by value --------
    def delete(self, value):
        if self.head is None:
            print("LIST IS EMPTY")
            return

        # If head node is to be deleted
        if self.head.data == value:
            self.head = self.head.next
            return

        temp = self.head
        while temp.next and temp.next.data != value:
            temp = temp.next

        if temp.next is None:
            print("Value not found")
            return

        temp.next = temp.next.next

        # -------- Delete by index (0-based) --------
    def delete_index(self, index):
        if self.head is None:
            print("LIST IS EMPTY")
            return

        if index < 0:
            print("Invalid index")
            return

        # Case 1: delete head node
        if index == 0:
            self.head = self.head.next
            return

        temp = self.head
        for _ in range(index - 1):
            if temp.next is None:
                print("Index out of range")
                return
            temp = temp.next

        if temp.next is None:
            print("Index out of range")
            return

        temp.next = temp.next.next

````

## 6. Searching in Linked List


````python

def search(head, key):
    pos = 1
    temp = head
    while temp:
        if temp.data == key:
            return pos
        temp = temp.next
        pos += 1
    return -1
````


## GATE PYQ

## 7. Reversing a Linked List (VERY IMPORTANT FOR GATE)


````python

def reverse(head):
    prev = None
    curr = head
    
    while curr:
        nxt = curr.next
        curr.next = prev
        prev = curr
        curr = nxt
    
    return prev
````

## 8. Detect Loop in Linked List (Floyd‚Äôs Cycle Detection)


````python

def detect_loop(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False
````

It checks whether a linked list contains a loop (cycle) or not.

10 ‚Üí 20 ‚Üí 30 ‚Üí 40
          ‚Üë      ‚Üì
          60 ‚Üê 50


slow = fast = head

Meaning:

slow moves 1 step at a time

fast moves 2 steps at a time

Both start at the head

This is why it‚Äôs also called Tortoise and Hare algorithm.

while fast and fast.next:

Why this condition?

fast should not be NULL

fast.next should not be NULL

Otherwise:

fast.next.next would cause an error

It also means no loop exists

üëâ GATE trap: If fast reaches NULL, there is no cycle.

fast = fast.next.next

Fast pointer moves 2 nodes

if slow == fast:
    return True

Why does this work?

If there is a loop:

Fast pointer will eventually lap slow

Just like a faster runner on a circular track

So:

If slow and fast meet ‚Üí loop exists

Time & Space Complexity (GATE favorite)

Time: O(n)

Space: O(1) (no extra memory)

GATE SUMMARY (Memorize)

Floyd‚Äôs algorithm = O(n), O(1)

Fast = 2 steps, Slow = 1 step

Meeting ‚áí loop exists

Meeting point ‚â† start of loop

Reset slow to head to find loop start after meeting.

## 11. Circular Linked List

A Circular Linked List (CLL) is a linked list in which:

The last node points back to the first node instead of pointing to NULL.


---

EXAPLES:

10 ‚Üí 20 ‚Üí 30 ‚Üí NULL



10 ‚Üí 20 ‚Üí 30
‚Üë           ‚Üì
‚Üê ‚Üê ‚Üê ‚Üê ‚Üê ‚Üê



‚úî No NULL pointer
‚úî Traversal is cyclic

---

Why Circular Linked List is needed?

Advantages:

1. No NULL checks at end

2. Useful in round-robin scheduling

3. Easy rotation of list

4. Continuous traversal

GATE relevance: Mostly conceptual + pointer dry run

---

Representation using head

1. We keep only one pointer:

    head ‚Üí first node

2. Last node:

    last.next = head


````python
class NODE:
    def __init__(self, data):
        self.data = data
        self.next = None


class CLL:
    def __init__(self):
        self.head = None

    # -------- Traversal --------
    def traversal(self):
        if self.head is None:
            print("LIST IS EMPTY")
            return

        temp = self.head               # ‚ùå was: temp = head
        while True:
            print(temp.data, end=" -> ")
            temp = temp.next
            if temp == self.head:      # loop termination condition
                break
        print("(back to head)")

    # -------- Length --------
    def cll_length(self):
        if self.head is None:
            return 0

        temp = self.head               # ‚ùå was: temp = head
        count = 1                      # start with 1 because head exists
        while temp.next != self.head:
            temp = temp.next
            count += 1
        return count

    # -------- Insert at Beginning --------
    def prepend(self, element):
        new_node = NODE(element)

        if self.head is None:
            new_node.next = new_node   # single node circular link
            self.head = new_node       # ‚ùå missing
            return

        temp = self.head
        while temp.next != self.head:  # find last node
            temp = temp.next

        new_node.next = self.head
        temp.next = new_node
        self.head = new_node           # ‚ùå was using local head
        return

    # -------- Insert at End --------
    def append(self, element):
        new_node = NODE(element)

        if self.head is None:
            new_node.next = new_node
            self.head = new_node
            return

        temp = self.head
        while temp.next != self.head:
            temp = temp.next

        temp.next = new_node
        new_node.next = self.head
        return

    # -------- Insert at Index --------
    def insert(self, element, index):
        length = self.cll_length()

        if index < 0 or index > length:
            print("INDEX OUT OF BOUND")
            return

        if index == 0:
            self.prepend(element)
            return

        temp = self.head
        for _ in range(index - 1):     # ‚ùå was: while index (infinite loop)
            temp = temp.next

        new_node = NODE(element)
        new_node.next = temp.next
        temp.next = new_node
        return

    # -------- Delete from Beginning --------
    def delete_begin(self):
        if self.head is None:
            return

        if self.head.next == self.head:
            self.head = None           # ‚ùå missing update
            return

        temp = self.head
        while temp.next != self.head:
            temp = temp.next

        self.head = self.head.next
        temp.next = self.head
        return

    # -------- Delete from End --------
    def delete_end(self):
        if self.head is None:
            return

        if self.head.next == self.head:
            self.head = None
            return

        prev = None
        curr = self.head
        while curr.next != self.head:
            prev = curr
            curr = curr.next

        prev.next = self.head
        return

    # -------- Delete by Index (NEW) --------
    def delete_at_index(self, index):
        length = self.cll_length()

        if index < 0 or index >= length:
            print("INDEX OUT OF BOUND")
            return

        if index == 0:
            self.delete_begin()
            return

        temp = self.head
        for _ in range(index - 1):
            temp = temp.next

        temp.next = temp.next.next
        return
````

## 13. Reverse Linked List (Recursive)


````python

def reverse_recursive(head):
    if head is None or head.next is None:
        return head
    
    new_head = reverse_recursive(head.next)
    head.next.next = head
    head.next = None
    return new_head
````


### GATE PYQ Concept
**Q:** What is the extra space used in recursive reversal?
**Answer:** O(n) due to recursion stack



## 14. Find Middle of Linked List
(Fast & Slow Pointer)


````python

def find_middle(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow.data
````


## 15. Nth Node from End


````python

def nth_from_end(head, n):
    first = second = head
    for _ in range(n):
        if first is None:
            return -1
        first = first.next
    
    while first:
        first = first.next
        second = second.next
    return second.data
````


## 16. Detect and Remove Loop


````python

def remove_loop(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            break
    else:
        return head
    
    slow = head
    while slow.next != fast.next:
        slow = slow.next
        fast = fast.next
    fast.next = None
    return head
````


## 17. Intersection Point of Two Linked Lists


````python

def get_intersection(head1, head2):
    p1, p2 = head1, head2
    while p1 != p2:
        p1 = p1.next if p1 else head2
        p2 = p2.next if p2 else head1
    return p1.data if p1 else -1
````


## 18. Check Palindrome Linked List


````python

def is_palindrome(head):
    slow = fast = head
    stack = []
    
    while fast and fast.next:
        stack.append(slow.data)
        slow = slow.next
        fast = fast.next.next
    
    if fast:
        slow = slow.next
    
    while slow:
        if stack.pop() != slow.data:
            return False
        slow = slow.next
    return True
````


## 19. Merge Two Sorted Linked Lists


````python

def merge_sorted(l1, l2):
    dummy = Node(0)
    tail = dummy
    
    while l1 and l2:
        if l1.data < l2.data:
            tail.next = l1
            l1 = l1.next
        else:
            tail.next = l2
            l2 = l2.next
        tail = tail.next
    
    tail.next = l1 if l1 else l2
    return dummy.next
````
